```python
import numpy as np
from numpy import fft
import matplotlib.pylab as plt

from blusky.wavelets.wavelet_factories_1d import vanilla_gabor_1d, vanilla_morlet_1d
```

# Demo wavelet factories
Use the factory functions to produce sets of Morlet1D wavelets at different resolutions, and also generate Morlet2D wavelets with the definitions used in https://github.com/scatnet/scatnet.

```python
def do_vis(wav, sample_rate=0.001):
    f1 = fft.fftshift(fft.fft(wav))
    freqz = fft.fftshift(np.fft.fftfreq(len(wav), sample_rate))

    plt.figure()
    fig, axs = plt.subplots(2,2,figsize=(20,10))   
    axs[0,0].plot(wav.real, '-*')
    axs[0,1].plot(wav.imag, '-*')
    axs[1,0].plot(freqz, np.abs(f1), '-*')
    axs[1,1].plot(freqz, np.angle(f1), '-*')
    plt.show()
```

# Morlet wavelet factories

```python
wav1 = vanilla_morlet_1d(0.001, j=1, crop=10.).kernel()
wav2 = vanilla_morlet_1d(0.001, j=2, crop=10.).kernel()
wav3 = vanilla_morlet_1d(0.001, j=3, crop=10.).kernel()
wav4 = vanilla_morlet_1d(0.001, j=4, crop=10.).kernel()
wav5 = vanilla_morlet_1d(0.001, j=5, crop=10.).kernel()

do_vis(wav2, sample_rate=0.001)
```

# Gabor wavelet factories

```python
wav1 = vanilla_gabor_1d(0.001, j=1).kernel(shape=128)
wav2 = vanilla_gabor_1d(0.001, j=2).kernel(shape=128)
wav3 = vanilla_gabor_1d(0.001, j=3).kernel(shape=128)
wav4 = vanilla_gabor_1d(0.001, j=4).kernel(shape=128)
wav5 = vanilla_gabor_1d(0.001, j=5).kernel(shape=128)

do_vis(wav1, sample_rate=0.001)
do_vis(wav5, sample_rate=0.001)
```
